<font size="+1">**課題②**</font><br>
$y=\sin(x)$のグラフを$[-\pi, \pi]$の区間で描画することを考えます．<br>
<br>
**課題②-1**<br>
講義スライドに倣い，numpyとmatplotlibのpyplotをインポートしましょう．

```
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

**課題②-2**<br>
$[-5, 5]$の区間で0.01刻みの小数列を要素に持つ配列<code>x</code>を生成しましょう．<br>
また，配列<code>x</code>の各要素に対して$y=\sin2x$を通した数を要素に持つ配列<code>y</code>を生成しましょう．

```
x = np.arange(-5, 5, 0.01)
y = np.sin(2*x)
```

**課題②-3**<br>
生成した配列<code>x</code>および<code>y</code>を用いて$y=\sin2x$のグラフを描画し，表示しましょう．<br>
この段階ではタイトルや範囲などを指定する必要はありません．

```
plt.plot(x, y)
```

**課題②-4**<br>
課題②-3のコードを改変して，x軸方向の表示範囲を$[-\pi, \pi]$にしましょう．

```
plt.xlim(-np.pi, np.pi)
plt.plot(x, y)
```

**課題②-5**<br>
課題②-4のコードを改変して，タイトルを付けて表示しましょう．

```
# 日本語を使用する場合
# !pip install japanize-matplotlib
# import japanize_matplotlib

plt.xlim(-np.pi, np.pi)
plt.title('Plot of y = Sin2x')
plt.plot(x, y)
```

**課題③**<br>
以下の条件を満たすグラフを描画し，表示してください．<br>
ただし，x軸の要素を表す変数<code>x</code>は**課題②-2**と同じような形式で定義域を拡張したものを用いてください．

* 関数 $y=\exp x \times \sin4x$ を $[-2\pi,2\pi]$ の定義域で表示
* x軸，y軸の両方に文字列「x」「y」のラベルを表示
* グラフは罫線付きで表示

```
# 関数の実装
# 先に定義したx, yと混同することを避けるため、ここでの配列はx_2, y_2と定義する。
x_2 = np.arange(-2*np.pi, 2*np.pi, 0.01)
y_2 = np.exp(x_2) * np.sin(4*x_2)

# 定義域の実装
plt.xlim(-2*np.pi, 2*np.pi)

# ラベルの実装
plt.xlabel('x')
plt.ylabel('y')

# グラフの罫線を表示
plt.grid()

# グラフを描画
plt.plot(x_2, y_2)
```
