**課題②** \\
以下のコードは，CSVファイルから読み込んだデータをグラフに描画するプログラムです．<br>
グラフの横軸を年，縦軸を平均寿命とし男性の平均寿命をo，女性の平均寿命を*で描画してください．<br>
コード内の<code>[ウ]〜[キ]</code>を埋め適切に動作するようにしてください． \\
ただし，<code>[ア][イ]</code>には課題①と同じものが入ります．

```
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("https://bit.ly/3707GMw")
a = data.values

x =  a[:, 0] # Year
y1 = a[:, 1] # Male
y2 = a[:, 2] # Female

#データ点の描画
plt.plot(x, y1, linewidth=0, marker="o", markersize=5) # Male
plt.plot(x, y2, linewidth=0, marker="*", markersize=5) # Female
```

**課題③** \\
以下のコードは，sklearnの最小２乗法を適用し、課題②で作成したグラフに追加で描画するプログラムです． \\
男性，女性それぞれで最小2乗法を行ない，直線を描画するプログラムを記述してください． \\
ただし，<code>[ア]～[キ]</code>には課題①②と同様のものが入ります．

```
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

data = pd.read_csv("https://bit.ly/3707GMw")
a = data.values

x =  a[:, 0] # Year
y1 = a[:, 1] # Male
y2 = a[:, 2] # Female

#データ点の描画
plt.plot(x, y1, linewidth=0, marker="o", markersize=5) # Male
plt.plot(x, y2, linewidth=0, marker="*", markersize=5) # Female

#以下に男性、女性それぞれの平均寿命に対して最小２乗法を適用するプログラムを記述
def hoge(x, y): # 繰り返しの動作を関数化してまとめる
    Ir = linear_model.LinearRegression()

    x = x.reshape(-1, 1)
    y = y.reshape(-1, 1)
    Ir.fit(x, y)

    yy = Ir.predict(x)
    return yy

yy1 = hoge(x, y1)
plt.plot(x, yy1)

yy2 = hoge(x, y2)
plt.plot(x, yy2)
```
